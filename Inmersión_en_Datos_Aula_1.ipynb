### Solución personal

Primero se crea una variable inmuebles_por_barrio donde se cálcula el promedio de las áreas de los inmuebles y se agrupa por barrios de forma descendente

```python
inmuebles_por_barrio = inmuebles.groupby('Barrio')['Area'].mean().sort_values(ascending=False)
inmuebles_por_barrio
```

Se crea una verificación sobre el promedio de área específicamente en el barrio Modelia y se compara con el anterior

```python
modelia = (inmuebles.Barrio == 'Modelia')
prom_area_modelia = inmuebles[modelia].Area.mean()
prom_area_modelia
```

```python
#Se importa librería para crear los gráficos
import matplotlib.pyplot as plt

inmuebles_por_barrio.head(10).plot.bar()

#Se personaliza la gráfica
plt.title('Promedio Area por Barrio')
plt.xlabel('Barrio')
plt.ylabel('Prom Area')

#Se muestra la grafica
plt.show()
```

```python
#Se calcula la mediana
mediana_area_barrio = inmuebles_por_barrio.median()
mediana_area_barrio
```

```python
#Las áreas mínimas y máximas existente de todos los barrios
min_area_barrio = inmuebles_por_barrio.min()
print('El area más pequeña es', min_area_barrio)

max_area_barrio = inmuebles_por_barrio.max()
print('El area más grande es', max_area_barrio)
```

Muchas veces para tratar datos se utiliza la función **.split** la cual lo que hace es separar una cadena de texto en función de un parámetro que sea suministrado, en caso de no hacerlo, por default toma los espacios para separar la cadena y devuelve una lista con los valores separados

```python
inmuebles.Valor[0].split()
```

El **inmuebles.Valor** por sí solo es un panda series, así que tomamos esta información y la convertimos en un string para poderla separar

```python
type(inmuebles.Valor)
```

```python
type(inmuebles.Valor.str)
```

```python
inmuebles.Valor.str.split()
```

```python
type(inmuebles.Valor.str.split())
```

Vemos que al aplicar la función .split tambien nos devuelve un panda series. Nos puede servir utilizarlo mejor en formato dataframe, para ello se hace

```python
inmuebles.Valor.str.split(expand = True)
```

Sin el parametro expands=True devuelve una columna de tipo panda series; cuando se coloca devuelve un dataframe de dos columnas unicamente con esa información

Es útil ya que se puede generar nuevas columnas para nuestro dataset para facilitar el tratamiento de los datos

```python
valor = inmuebles.Valor.str.split(expand = True) #Asignamos a una variable el resultado del dataset
inmuebles['Moneda'] = valor[0] #Al dataset inmuebles añadimos una nueva columna llamada Moneda a la cual le asignamos los valores de la columna 0 del dataset nuevo
inmuebles['Precio'] = valor[1] #Lo mismo hacemos con el precio
inmuebles.sample(3)
```

```python
inmuebles.info()
```

Nuestra columna precio aún sigue siendo objeto, aún no es un valor númerico. Debemos hacer un tratamiento adicional ya que el punto aunque es un separador de miles, el programa no lo lee igual, se debe convertir

```python
inmuebles['Precio'].sum() #Vemos que concatena, no hay ninguna operación aritmética
```

```python
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.', '', regex = True)
#Primero realizamos el tratamiento en strings y despues usamos la función replace que nos permite reemplazar dos valores
#El primer parámetro es el valor que se quiere reemplazar. El segundo parámetro es por cuál se quiere reemplazar.
#El tercer parámetro se debe escribir regex=True
```

```python
inmuebles[['Precio', 'Barrio']] #Se escribe entre dos llaves para indicarle a python que se quieren en formato dataframe
```

```python
inmuebles.info()
```

Precio sigue siendo un objeto, ésto debido a que solo eliminamos el separador de miles, es decir, el punto. Ahora debemos convertirlo a un tipo de dato númerico

```python
inmuebles['Precio_Millon'] = inmuebles.Precio.astype('float')/1000000
#La función astype nos permite convertirlo a un tipo float
#Se divide por un millón para simplificar los valores, ya que todo se maneja en millones, dentro del dataset. Permite mejor lectura y visualización
inmuebles.info()
```

El histograma ayuda a entender la disfritubicón de frecuencias de una variable que en este caso es el precio de los inmuebles. Los bins son la separación de los rangos bloques que vamos a visualizar en nuestro eje horizontal

```python
import seaborn as sns
```

```python
plt.figure(figsize = (10, 6)) #Permite configurar el tamaño para las gráficas
grafica = sns.histplot(data = inmuebles, x = 'Precio_Millon', kde = True, hue = 'Tipo')
#Permite crear un gráfico de histograma
#Parametro, data es de donde salen los datos
#Parametro x, se estipula cual será el eje x y sus valores
#Parametro kde, es la curva de densidad
grafica.set_title('Distribución de Valores de inmuebles en Bogotá') #Asigna un titulo a la grafica
plt.xlim(50, 1000) #Limita el rango de valores en el eje x
plt.savefig('/content/drive/MyDrive/inmersion_datos/valor_inmuebles.png', format = 'png')
plt.show()
```
